```python
import pandas as pd # importing the necessary libraries
import numpy as np
```

```python
df = pd.read_csv('D:\My files\Filtered_Bulleting_Gold.csv') # importing csv file
```

```python
pd.set_option('display.max_columns', 8) # displaying maximum columns and rows
pd.set_option('display.max_rows', 447)

df.replace("NaN", np.nan, inplace=True) #handling missing data
df.replace("None", np.nan, inplace=True)
```

```python
df['Reference_Date'] = pd.to_datetime(df['Reference_Date']) # converting reference_date into datetime format
```

```python
df.set_index('Reference_Date') #set reference_date as index
```

```python
df.describe()
```

```python
df[['Gold_Ask_995','Gold_Bid_995']].describe() #basic statistics about specific columns
```

```python
df['MA30'] = df['Gold_Ask_995'].rolling(30).mean()
```

```python
df['STD30'] = df['Gold_Ask_995'].rolling(30).std()
```

```python
df # dataframe with 30 day window rolling moving average and standard deviation
```

```python
corr_df = df[['Gold_Ask_995','Gold_Bid_995']]

corr_df.corr()
```

```python
import matplotlib.pyplot as plt


date = df['Reference_Date']
rolling_mean = df['MA30']
std_dev = df['STD30']
gold_ask = df['Gold_Ask_995']
gold_bid = df['Gold_Bid_995']

fig = plt.figure(figsize=(14, 8))

plt.plot(date,rolling_mean, label= '30-Day Rolling Mean (Ask)')
plt.plot(date,gold_ask, label= 'Gold Ask Price 995')
plt.plot(date,gold_bid , label ='Gold Bid Price 995')
plt.fill_between(date, 
                 df['MA30'] - df['STD30'], 
                 df['MA30'] + df['STD30'], 
                 color='green', alpha=0.2, label='30-Day Rolling Std Dev (Ask)')





plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Time Series of Gold Ask and Bid Prices (995) with Rolling Mean and Std Dev")

plt.grid(True)
plt.legend()
plt.show()
```
